```python
import pandas as pd
import numpy as np
# Import necessary libraries
import warnings
warnings.filterwarnings('ignore')  # Suppress warnings
```

```python
df=pd.read_excel('/train.xlsx')
df.head()
```

```python
df.info()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=["target"])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

# Add cluster labels to the dataframe
df["cluster"] = kmeans.labels_
print(df)
```

```python
def predict_cluster(data_point):
    # Preprocess the data point
    data_point_scaled = scaler.transform([data_point])
    # Predict the cluster
    cluster = kmeans.predict(data_point_scaled)
    return cluster[0]

# Example data point
data_point = [-70, -61, -66, -53, -51, -63, -82, -57, -76, -78, -66, -66, -61, -59, -73, -75, -63, -77]
cluster = predict_cluster(data_point)
print(f"The data point belongs to cluster {cluster}")
```
